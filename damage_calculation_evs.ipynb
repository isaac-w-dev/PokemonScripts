```python
def calc_health_from_total(total, iv, ev, level=50):
    base_health = (100*(total-10-level)/level - iv - ev//4)/2
    return base_health
```

```python
def calc_base_from_total(total, iv, ev, nature = 1, level=50):
    base_stat_float = ((total/nature - total/nature % 1 - 5) * 100/level - iv - ev // 4)/2
    return base_stat_float
```

```python
def calc_total_from_base(base, nature, ev, iv = 31, level=50):
    total = ((iv + 2 * base + ev // 4) * level/100 + 5) * nature // 1
    return total
```
