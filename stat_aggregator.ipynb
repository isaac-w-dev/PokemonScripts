```python
import os
import re
from pokemon import pokemon
# import pytest
```

```python

def write_pkmn_to_file(pkmn_list):
    with open("pkmn_list.py", "w" ) as f:
        f.write("from pokemon import pokemon\n\n")
        f.write(f"pk_list = [\n\tpokemon('{pkmn_list[0].tier}', '{pkmn_list[0].name}', {pkmn_list[0].types}, {pkmn_list[0].abilities},{pkmn_list[0].HP}, {pkmn_list[0].Atk}, {pkmn_list[0].Def}, {pkmn_list[0].SpAtk}, {pkmn_list[0].SpDef}, {pkmn_list[0].Spd}, {pkmn_list[0].BST})")
        for pkmn in pkmn_list[1:]:
            f.write(f",\n\tpokemon('{pkmn.tier}', '{pkmn.name}', {pkmn.types}, {pkmn.abilities}, {pkmn.HP}, {pkmn.Atk}, {pkmn.Def}, {pkmn.SpAtk}, {pkmn.SpDef}, {pkmn.Spd}, {pkmn.BST})")
        f.write("\n]")
```

```python
def parse_viablility(pkmn_array):
    with open("viable_pokemon.txt") as vp:
        vp_array = vp.read().split('\n')
        pkmn_array = [pkmn for pkmn in pkmn_array if any(item in pkmn.name for item in vp_array)]
    return pkmn_array
```

```python
def separate_pokemon(pkmn_list):
    return re.split("(BST\n.{3}\n)", pkmn_list)
```

```python
def append_BST(pkmn_list):
    new_list = []
    for i in range(0, len(pkmn_list[:]) - 1 , 2):
        new_list.append(pkmn_list[i] + pkmn_list[i + 1])
    return new_list
```

```python
def filter_tier(unfiltered_tier):
    tier = ''
    for char in unfiltered_tier:
        if char == '(' or char == ')':
            continue
        tier += char
    return tier
```

```python
def separate_types(types):
    type_list = []
    type = types[0]
    for char in types[1:]:
        if char.isupper():
            type_list.append(type)
            type = ''
        type += char
    type_list.append(type)
    return type_list
```

```python
def separate_abilities(ability_string):
    ability_list = []
    ability = ability_string[0]
    spaced = False

    for char in ability_string[1:]:
        if char in '\n()': continue
        if char.isupper() == True:
            if spaced == False:
                ability_list.append(ability)
                ability = ''
        spaced = False
        if char == ' ' or char == '-':
            spaced = True
        ability += char
    ability_list.append(ability)
    return ability_list
```

```python
def extract_abilities_bst(remaining_string):
    info_array = re.split('(HP)', remaining_string)
    info_array = [info_array[0], info_array[1]+info_array[2]]
    return info_array
```

```python
def check_bst(base_stat_array):
    total = 0
    for item in base_stat_array[:-1]:
        total += item
    assert total == base_stat_array[-1]
```

```python
def separate_base_stats(bst_string):
    bst_string = bst_string.strip()
    base_stat_array = re.split(' ?[A-Za-z]+\n', bst_string)
    base_stat_array.pop(0)
    base_stat_array = [int(item) for item in base_stat_array[:]]
    
    check_bst(base_stat_array)
    return base_stat_array
```

```python
def display_pkmn_from_list(pkmn_object_array):
    for pkmn in pkmn_object_array:
            print("Tier:", pkmn.tier, "\nPokemon:", pkmn.name, "\nTypes:", pkmn.types, "\nAbilities:", pkmn.abilities)
            print(pkmn.HP, pkmn.Atk, pkmn.Def, pkmn.SpAtk, pkmn.SpDef, pkmn.Spd, pkmn.BST)
```

```python
with open("AllPokemonInfo.txt") as f:
    pkmn_object_array = []
    raw_pkmn_list = separate_pokemon(f.read())
    pkmn_list = append_BST(raw_pkmn_list)
    for item in pkmn_list:
        tier, name, types, remaining_string = re.split(" ", item, maxsplit=3)
        tier = filter_tier(tier)
        types = separate_types(types)
        ability_string, bst_string = extract_abilities_bst(remaining_string)
        abilities = separate_abilities(ability_string)
        HP, Atk, Def, SpAtk, SpDef, Spd, BST = separate_base_stats(bst_string)
        pkmn_object_array.append(pokemon(tier, name, types, abilities, HP, Atk, Def, SpAtk, SpDef, Spd, BST))
    pkmn_object_array = parse_viablility(pkmn_object_array)
    write_pkmn_to_file(pkmn_object_array)
```
